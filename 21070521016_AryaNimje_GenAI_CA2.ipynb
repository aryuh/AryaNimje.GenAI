```python
#Q3 Generate a model for an Insurance company to hold information on the insurer's vehicle,
#and create a chart of monthly, yearly, and qtrly premiums based on no. of years of insurance
#where in each year, the value of the vehicle depreciates by 7%.

#Answer:
import matplotlib.pyplot as plt

class VehicleInsurance:
    def __init__(self, vehicle_value, years_of_insurance, depreciation_rate=0.07, premium_rate=0.05):
        self.vehicle_value = vehicle_value
        self.years_of_insurance = years_of_insurance
        self.depreciation_rate = depreciation_rate
        self.premium_rate = premium_rate
        self.yearly_premiums = []
        self.quarterly_premiums = []
        self.monthly_premiums = []

    def calculate_premiums(self):
        current_value = self.vehicle_value
        for year in range(self.years_of_insurance):
            yearly_premium = current_value * self.premium_rate
            self.yearly_premiums.append(yearly_premium)
            self.quarterly_premiums.append(yearly_premium / 4)
            self.monthly_premiums.append(yearly_premium / 12)
            current_value -= current_value * self.depreciation_rate  # Depreciating vehicle value by 7%

    def display_premiums(self):
        years = list(range(1, self.years_of_insurance + 1))

        plt.figure(figsize=(10, 6))
        plt.plot(years, self.yearly_premiums, label='Yearly Premium', marker='o')
        plt.plot(years, self.quarterly_premiums, label='Quarterly Premium', marker='o')
        plt.plot(years, self.monthly_premiums, label='Monthly Premium', marker='o')
        plt.xlabel('Years')
        plt.ylabel('Premium Amount')
        plt.title('Premiums over the Insurance Period')
        plt.legend()
        plt.grid(True)
        plt.show()


# Taking input from the user
vehicle_value = float(input("Enter the initial value of the vehicle: "))
years_of_insurance = int(input("Enter the number of years for insurance: "))

# Create the insurance object with the provided inputs
insurance = VehicleInsurance(vehicle_value, years_of_insurance)
insurance.calculate_premiums()
insurance.display_premiums()
```

```python
#Q4 Generate a model to represent interest calculations of a Bank account where the process of
#calculating interest for 6 months is a. Find minimum balance for each month b. Make a total of
#all minimum balances c. Calculate interest based on interest rate d. Divide interest by 12 to
#find one-month interest e. Multiply interest by 6 to show interest in the account. Generate a
#model to represent transactions and interest calculations for 6 months.

#Answer:
import random

class BankAccountWithInterest:
    def __init__(self, account_number, initial_balance, interest_rate):
        self.account_number = account_number
        self.balance = initial_balance
        self.transactions = []
        self.interest_rate = interest_rate

    def deposit(self, amount):
        self.transactions.append(f"Deposit: {amount}")
        self.balance += amount

    def withdraw(self, amount):
        if amount <= self.balance:
            self.transactions.append(f"Withdrawal: {amount}")
            self.balance -= amount
        else:
            self.transactions.append(f"Failed Withdrawal: {amount}")

    def calculate_minimum_balance(self, months):
        min_balances = []
        for month in range(1, months + 1):
            # Simulating monthly transactions and getting a minimum balance (for the sake of the example, we use random values)
            monthly_min_balance = self.balance - random.randint(100, 500)
            min_balances.append(monthly_min_balance if monthly_min_balance > 0 else 0)
        return min_balances

    def calculate_interest(self, months):
        min_balances = self.calculate_minimum_balance(months)
        total_min_balances = sum(min_balances)
        interest_for_6_months = (total_min_balances / months) * (self.interest_rate / 100)
        one_month_interest = interest_for_6_months / 12
        six_month_interest = one_month_interest * 6
        return six_month_interest


# Taking user input
account_number = input("Enter the account number: ")
initial_balance = float(input("Enter the initial balance of the account: "))
interest_rate = float(input("Enter the interest rate (in percentage): "))

account = BankAccountWithInterest(account_number, initial_balance, interest_rate)

interest = account.calculate_interest(6)

print(f"Interest for 6 months: {interest:.2f}")


```
